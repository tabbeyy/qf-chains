```python
import pandas as pd
```

In this example we're using pandas to load the CSV files containing our option chains.
If you haven't used pandas before you can learn more [here](https://pandas.pydata.org/docs/user_guide/10min.html).

```python
def load_chains(ticker: str) -> tuple[pd.DataFrame, pd.DataFrame]:
    calls_df = pd.read_csv(f"{ticker}_calls.csv")
    puts_df = pd.read_csv(f"{ticker}_puts.csv")
    
    return calls_df, puts_df
```

```python
calls_df, puts_df = load_chains(ticker="AAPL")
```
